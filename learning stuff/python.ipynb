```python
def someFunc(bonus):
    return bonus * 2

someList = [100, 200, 300]

iterator = map(someFunc, someList) # map() applies a function to every item of iterable, yielding the results.

print(list(iterator))
```

```python
scores = [70, 60, 80, 90, 50]
filtered = filter(lambda score: score >= 70, scores) # filter() constructs an iterator from those elements of iterable for which function returns true.

print(list(filtered))
```

```python
s1 = {'Python', 'Java', 'C++'} # This is a set! not a tuple or dictionary
s2 = {'C#', 'Java', 'C++'}

s = s1.intersection(s2) # intersection() returns a new set with elements common to the set and all others.

print(s)

s = s1 & s2 # & operator is equivalent to intersection()

print(s)

s = s1.difference(s2) # difference() returns a new set with elements in the set that are not in the others.

print(s)

s = s1 - s2 # - operator is equivalent to difference()

print(s)

s = s1.symmetric_difference(s2) # symmetric_difference() returns a new set with elements in either the set or other but not both.

print(s)
```

```python
from typing import Union

number = Union[int, float]

def add(x: number, y: number) -> number:
    return x + y
```

```python
try:
    # get input net sales
    print('Enter the net sales for')

    previous = float(input('- Prior period:'))
    current = float(input('- Current period:'))

    # calculate the change in percentage
    change = (current - previous) * 100 / previous

    # show the result
    if change > 0:
        result = f'Sales increase {abs(change)}%'
    else:
        result = f'Sales decrease {abs(change)}%'

    print(result)
except ValueError:
    print('Error! Please enter a number for net sales.')
except ZeroDivisionError:
    print('Error! The prior net sales cannot be zero.')

except:
    print('Error! Something went wrong.')

finally:
    print('Done.')
```
